```python
df.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country","region_txt":"Region","provstate":"Province/State","city":"City","latitude":"Latitude","longitude":"Longitude","location":"Location","summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type","gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type","nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"},inplace=True)
```

```python
df["Killed"]=df["Killed"].fillna(0)
df["Wounded"]=df["Wounded"].fillna(0)
df["Casualty"]=df["Killed"]+df["Wounded"]
```

```python
df.describe()
```

```python
#data Visualization
attacks=df["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head()
```

```python
attacks.plot(kind="bar",color="Red",figsize=(10,3),fontsize=10)
plt.title("Timeline of Attacks", fontsize=12)
plt.xlabel("Years",fontsize=12)
plt.ylabel("Number of Attacks",fontsize=12)
plt.show()
```

```python
yc=df[["Year","Casualty"]].groupby("Year").sum()
yc.head()
#yc = Year casualty
```

```python
yc.plot(kind="bar",color="Blue",figsize=(10,3))
plt.title("Year wise Casulties",fontsize=10)
plt.xlabel("Years",fontsize=10)
plt.xticks(fontsize=9)
plt.ylabel("Number of Casulties",fontsize=10)
plt.show()
```

```python
yk=df[["Year","Killed"]].groupby("Year").sum()
yk.head()
```

```python
yk.plot(kind="bar",color="blue",figsize=(10,3))
plt.title("People Killed in each Year")
plt.xlabel("Years")
plt.ylabel("Number of People Killed")
plt.show()
```

```python
yw=df[["Year","Wounded"]].groupby("Year").sum()
yw.head()
```

```python
yw.plot(kind="bar",color="Blue",figsize=(10,3))
plt.title("People wounded in each Year")
plt.xlabel("Years")
plt.ylabel("Number of People Wounded")
plt.show()
```

```python
reg=pd.crosstab(df.Year,df.Region)
reg.head()
```

```python
reg.plot(kind="area",stacked=False, alpha=0.5,figsize=(20,10))
plt.title("Region Wise Attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(15,7))
plt.title("Total Number of Attacks in Region From 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
rc=df[["Region","Casualty"]].groupby("Region").sum().sort_values(by="Casualty",ascending=False)
rc
```

```python
rc.plot(kind="bar",color="Blue",figsize=(15,6))
plt.title("Region Wise Casualty",fontsize=13)
plt.xlabel("Regions",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Numbers of Casualty",fontsize=13)
plt.show()
```
